# Getting Zip Codes from Latitude and Longitude

```python
import geopy
```

```python
df1 = df5_cleaned[["start_lat", "start_lng"]]
df1
```

```python
df2 = df5_cleaned[["end_lat", "end_lng"]]
df2
```

```python
df1.to_csv("start_coord.csv")
df2.to_csv("end_coord.csv")
```

```python
geolocator = geopy.Nominatim(user_agent="http")
```

```python
def get_zip_code(x, lat, lng):
    location = geolocator.reverse((x[lat],x[lng]))
    try:
        return location.raw['address']['postcode']
    except Exception as e:
        print(e)
```

```python
df1['start_zipcode'] = df1.head().apply(lambda x: get_zip_code(x, 'start_lat', 'start_lng'), axis = 1)
df1
```

```python
df1_section = df1.iloc[:100]
df1_section
```

```python
# Takes ~50s to execute!
df1_section['start_zipcode'] = df1_section.apply(lambda x: get_zip_code(x, 'start_lat', 'start_lng'), axis = 1)
df1_section
```

```python
df2['end_zipcode'] = df2.head().apply(lambda x: get_zip_code(x, 'end_lat', 'end_lng'), axis = 1)
df2
```
